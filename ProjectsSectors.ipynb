```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('Projects.csv')

df.columns = df.columns.str.strip()

df['Start'] = pd.to_numeric(df['Start'], errors='coerce')
df['End'] = pd.to_numeric(df['End'], errors='coerce')

df = df.dropna(subset=['Start', 'End'])

rows = []
for _, row in df.iterrows():
    for year in range(int(row['Start']), int(row['End']) + 1):
        rows.append({'Year': year, 'Problem to address': row['Problem to address']})

expanded_df = pd.DataFrame(rows)

grouped = expanded_df.groupby(['Year', 'Problem to address']).size().unstack(fill_value=0)

grouped.plot(kind='bar', stacked=True, figsize=(12, 8))

plt.title('Number of Projects per Year by Sector')
plt.xlabel('Year')
plt.ylabel('Number of Projects')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
df.columns = df.columns.str.strip()

# Ensure Start and End columns are of integer type
df['Start'] = pd.to_numeric(df['Start'], errors='coerce')
df['End'] = pd.to_numeric(df['End'], errors='coerce')

# Drop rows with invalid Start/End dates
df = df.dropna(subset=['Start', 'End'])

# Create a new DataFrame to expand each project for each year it was active
rows = []
for _, row in df.iterrows():
    for year in range(int(row['Start']), int(row['End']) + 1):
        rows.append({'Year': year, 'Problem to address': row['Problem to address']})

expanded_df = pd.DataFrame(rows)

# Group by year and 'Problem to address' and count the number of projects per year
grouped = expanded_df.groupby(['Year', 'Problem to address']).size().unstack(fill_value=0)

# Create a list of unique "Problem to address" categories
problems = grouped.columns

# Set up subplots with a 2x3 layout for better readability (2 rows and 3 columns)
fig, axes = plt.subplots(2, 3, figsize=(15, 10), sharey=True)

# Flatten the axes for easy iteration
axes = axes.flatten()

# Plot each problem category on a separate axis
for i, problem in enumerate(problems):
    grouped[problem].plot(kind='bar', ax=axes[i], width=0.8)
    axes[i].set_title(problem)
    axes[i].set_xlabel('Year')
    if i % 3 == 0:  # Set y-label for the first plot in each row
        axes[i].set_ylabel('Number of Projects')
    axes[i].set_xticklabels(axes[i].get_xticklabels(), rotation=45, ha='right')

# Remove the empty subplot if there are fewer than 6 categories
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```
