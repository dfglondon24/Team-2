```python
import pandas as pd
```

```python
def read_data(filename: str):
    """
    Function to read in an excel file, and convert it to a Pandas dataframe
    """
    
    print("Note: filename input should not include 'shared_data_read_only/'")
    print("Example: filename = 'AfriKids/InternalData/1. Master Stats 2023.xlsx'")
    
    df = pd.read_excel('shared_data_read_only/Data/' + filename)
    
    return df
```

```python
# Example use

# Read in file
df = read_data('AfriKids/InternalData/1. Master Stats 2023.xlsx')

# Print first 5 rows
df.head(5)
```
